## Prediction


```python
X = df1[['componente_1_CH', 'componente_1_PLA', 'componente_1_ST', 'componente_1_CMC']].copy()
Y = df1[['módulo_tensil_(gpa)']].copy()


print(X.shape)
print(Y.shape)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
from sklearn import cross_validation
from sklearn import metrics

seed = 7

model = LinearRegression()

predicted = cross_validation.cross_val_predict(model, X, Y, cv=2)
print("R^2:" + str(metrics.r2_score(Y, predicted))) 
print("RMSE:" + str(metrics.mean_squared_error(Y, predicted))) 
print("RMAE:" + str(metrics.mean_absolute_error(Y, predicted))) 
```
